```python
t=int(input())
```

```python
def AOPS(T1,T2,t):
    if len(T1)>=len(T2):
        q=T1
        p=T2
    else:
        q=T2
        p=T1    
    
    for i in range(0,len(q)-len(p)+1):
        S=q[i:len(p)+i]
        L=[]
        count=0
        for j in range(0,len(p)):
            if p[j]!=S[j]:
                L.append(j)
                count+=1
        if count<=t:
            print(i)
                
        
            
```

```python
AOPS(T1,T2,t)
```
